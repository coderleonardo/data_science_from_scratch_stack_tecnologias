# Regress√£o linear

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

## Leitura dos dados

```python
X = np.array([0, 1, 2, 3])
y = np.array([0, 1, 2, 3])
```

## Representando uma Reta

```python
def predict(alpha, beta, X):
    return alpha * X + beta
```

```python
alpha = 0.50 # Coeficiente angular da reta
beta = 0 # Coeficiente linear da reta 

y_hat = predict(alpha=alpha, beta=beta, X=X)

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X,y, c='b')
plt.plot(X, y_hat, 'r')
plt.text(0,2.90,'Alpha={} e Beta={}'.format(alpha, beta), fontdict={'size': 15})
plt.show()
```

## Escolhendo o melhor alpha e beta

```python
tamanho = len(X)
X_b = np.c_[np.ones((tamanho, 1)), X]  # add x0 = 1 to each instance

X_b
```

```python
# M√©todo dos m√≠nimos quadrados
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best # (linear, inclina√ß√£o)
```

```python
alpha = theta_best[1] # Inclina√ß√£o
beta = theta_best[0]
```

```python
y_hat = predict(alpha=alpha, beta=beta, X=X)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X,y, c='b')
plt.plot(X, y_hat, 'r')
plt.text(0,2.90,'Alpha={} e Beta={}'.format(round(alpha,3), round(beta,3)), fontdict={'size': 15})
```
